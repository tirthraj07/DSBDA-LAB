Convert the types into proper datatypes

```python
df["discounted_price"] = df["discounted_price"].astype(str).str.replace('₹',"")
df["discounted_price"] = df["discounted_price"].str.replace(",","")
df["discounted_price"] = df["discounted_price"].astype('float64')
```

```python
df["actual_price"] = df["actual_price"].astype(str)
df["actual_price"] = df["actual_price"].str.replace("₹",'')
df["actual_price"] = df["actual_price"].str.replace(",",'')
df["actual_price"] = df["actual_price"].astype('float64')
```

```python
df["discount_percentage"] = df["discount_percentage"].astype(str)
df["discount_percentage"] = df["discount_percentage"].str.replace("%","")
df["discount_percentage"] = df["discount_percentage"].astype("float64")
df["discount_percentage"] = df["discount_percentage"]/100
```

```python
df["rating"] = df["rating"].astype(str)
for x in df.index:
    if df.loc[x, "rating"] == "|":
        df.drop(x, inplace=True)

df["rating"] = df["rating"].astype('float64')
```

```python
df["rating_count"] = df["rating_count"].astype(str)
df["rating_count"] = df["rating_count"].str.replace(",","")
df["rating_count"] = df["rating_count"].astype("float64")
```

Handle Missing Values

You can drop the values using `dropna()`
Or you can forward fill the values using `ffill()`
It fills missing values (NaNs) by copying the last valid (non-NaN) value downward.

```python
df["rating_count"] = df["rating_count"].ffill()
```
