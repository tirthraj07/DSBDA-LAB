# Theory Related to Seaborn Before Assignment
Seaborn is a Python data visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. It simplifies the creation of complex visualizations like heatmaps, time series, categorical plots, regression plots, and more, using a simple syntax.

Seaborn offers a variety of plots to visualize data, especially statistical relationships. Here are some of the most common plot types you can create using Seaborn:

### 1. **Relational Plots**
   - **scatterplot**: Used to visualize the relationship between two continuous variables.
   - **lineplot**: Displays data points connected by lines, typically for time series or continuous data.
   - **relplot**: A general function that can create scatterplots or line plots, depending on the kind of data you provide.

### 2. **Categorical Plots**
   - **stripplot**: A scatter plot where one axis represents categories, useful for showing distribution within a category.
   - **swarmplot**: Similar to a strip plot but avoids overlap by adjusting the position of each point.
   - **boxplot**: Displays the distribution of a dataset through its quartiles and outliers.
   - **violinplot**: Combines aspects of boxplot and kernel density estimation, showing the distribution and density of data across categories.
   - **barplot**: Shows the distribution of data across categorical variables as bars.
   - **countplot**: Displays the count of observations in each category.
   - **pointplot**: Displays a point for each category, with optional error bars.

### 3. **Distribution Plots**
   - **histplot**: Creates histograms to show the distribution of a single variable.
   - **kdeplot**: Plots Kernel Density Estimation (KDE), which estimates the probability density function of a continuous random variable.
   - **distplot**: Combines histogram and KDE, though this is being replaced with `histplot` and `kdeplot` for better functionality.
   - **rugplot**: Displays small marks at the data points along an axis, often used to show the distribution of a variable.

### 4. **Matrix Plots**
   - **heatmap**: Visualizes a 2D matrix of data where individual values are represented by varying colors.
   - **clustermap**: A heatmap with hierarchical clustering of rows and/or columns.

### 5. **Regression Plots**
   - **regplot**: A scatter plot with a fitted regression line. It can also show the confidence intervals.
   - **lmplot**: Similar to `regplot` but with additional functionality for adding facets or grouping variables.
   - **residplot**: Displays the residuals of a regression model, useful for diagnostic purposes.

### 6. **Pairwise Plots**
   - **pairplot**: A matrix of scatter plots for all pairs of variables, useful for examining relationships between multiple variables at once.
   - **pairgrid**: A more customizable version of `pairplot`, giving finer control over individual plots in the grid.

### 7. **Facet Grids**
   - **facetgrid**: A way to create a grid of plots, divided by one or more categorical variables, enabling comparison of subsets of data.
   - **catplot**: A more modern, unified function for creating categorical plots with faceting.

### 8. **Other Plots**
   - **twinplot**: For displaying two types of plots (e.g., a bar plot and a line plot) on the same axes.
   - **jointplot**: Combines scatter plots and histograms/KDE for a joint view of two variables.

Seabornâ€™s flexibility and ease of use make it an excellent choice for visualizing data in many formats, especially for statistical insights!

# Commonly Used Plots

## 1. **Line Plot**
The lineplot in Seaborn is used to visualize the relationship between two continuous variables over a continuous axis, typically in the form of a line. It's especially useful for displaying trends or changes over time or any continuous variable.

## 2. **Scatter Plot**
A scatter plot is a type of data visualization that displays individual data points on a two-dimensional plane. It is used to represent the relationship between two continuous variables, with each point showing the values of those variables for a particular observation.

## 3. **Box Plot**
A box plot (also known as a box-and-whisker plot) is a data visualization that displays the distribution of a dataset through its quartiles, highlighting key statistical features such as the median, upper and lower quartiles, and potential outliers. It is commonly used to compare distributions across different categories or groups.

## 4. **Bar Plot**
A bar plot (or bar chart) is a type of data visualization that represents categorical data with rectangular bars. The length of each bar is proportional to the value of the variable it represents, making it useful for comparing quantities across different categories.

## 5. **Histogram**
A histogram is a type of data visualization that shows the distribution of a dataset by grouping data points into bins or intervals. It provides a way to visualize the frequency (or count) of observations within each bin, giving insights into the shape, spread, and central tendency of the data.

## 6. **KDE Plot**
A KDE plot (Kernel Density Estimate plot) is a data visualization that estimates the probability density function (PDF) of a continuous random variable. It smooths the data into a continuous curve, allowing you to understand the distribution of the data without the need for binning (as in a histogram). A KDE plot is useful for visualizing the shape and spread of the data, and it is often used to identify patterns such as skewness, multimodality (multiple peaks), or whether the data follows a normal distribution.

## 7. **Reg Plot**
A regplot is a type of plot in Seaborn that combines a scatter plot with a regression line. It is used to visualize the relationship between two continuous variables, allowing you to observe both the data points (scatter) and the best-fit regression line that represents the trend between those variables. The regression line can be linear (default) or can be customized for other types of fits.

````python
# Fill the missing values of Age with the median values
df["Age"] = df["Age"].fillna(df["Age"].median())
df.isna().sum()
````

2. Create a New Feature: "HasCabin"

````python
df["hasCabin"] = df["Cabin"].notna()
df = df.drop(columns=["Cabin"])
df
````

Dealing with Outliers

````python
def remove_outliers(df:pd.DataFrame, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    IQR = q3 - q1
    lower_bound = q1 - (1.5 * IQR)
    upper_bound = q3 + (1.5 * IQR)
    return df[(df[col]>=lower_bound) & (df[col]<=upper_bound)]
````

````python
df = remove_outliers(df, "Age")
df = remove_outliers(df, "Fare")
df.describe()
````

````python
df.isna().sum()
````

# Data Visualization

````python
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
# Count Plot to show how many people survived
sns.countplot(data=df, x="Pclass", hue="Survived")

plt.xlabel("Passenger Class (1 = First, 2 = Second, 3 = Third)")
plt.legend(title="Survival Status", labels=["Did Not Survive", "Survived"])
plt.title("Passenger Class vs Survival Count")
plt.show()
````

````python
# Plot Sex vs Survival

sns.countplot(data=df, x="Sex", hue="Survived")

plt.xlabel("Gender")
plt.legend(title="Survived Status",labels=["Not Survived","Survived"])
plt.title("Gender vs Survived")
plt.show()
````

````python
# hasCabin vs Survived
sns.countplot(data=df, x='hasCabin', hue='Survived')
plt.title('Survival Count Based on Cabin Availability')
plt.xlabel('Has Cabin (0 = No, 1 = Yes)')
plt.ylabel('Passenger Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
````

````python
# Calculate survival rate based on hasCabin
cabin_survival = df.groupby('hasCabin')['Survived'].mean().reset_index()

# Plot
sns.barplot(data=cabin_survival, x='hasCabin', y='Survived')
plt.title('Survival Rate Based on Having Cabin')
plt.xlabel('Has Cabin (0 = No, 1 = Yes)')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.show()
````

````python
# Age vs Fare
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived')
plt.title('Age vs Fare colored by Survival')
plt.show()
````

````python
#  Distribution of Fare according to survival

sns.kdeplot(data=df, x="Fare",hue='Survived', fill=True)
````

````python
# Age Distribution by Survival
sns.kdeplot(data=df, x='Age', hue='Survived', fill=True)
````

````python
# Boxplot for Fare distribution for Survived vs Not Survived
sns.boxplot(data=df, x='Survived', y='Fare')
plt.title('Fare Distribution by Survival Status')
plt.show()
````
